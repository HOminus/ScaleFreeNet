```julia
using Plots

struct Data
    degree::UInt32
    probability::Float64
    Data(d, p) = new(d, p)
end

degrees = Vector{Data}()

lines = readlines("dd/astro.txt")
for line in lines
    numbers = split(line, ' ')
    degree = parse(UInt32, numbers[1])
    prob = parse(Float64, numbers[2])
    if prob == 0 continue end
    push!(degrees, Data(degree, prob))
end

p = scatter([d.degree for d in degrees], [d.probability for d in degrees], yaxis=:log, xaxis=:log)
title!("Physics Collaboration Graph")
xlabel!("Degree k")
ylabel!("Probability p(k)")
display(p)
```
